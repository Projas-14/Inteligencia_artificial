# Análisis del dataset de Ventas en millones por trimestre

````python
shipments_in_mn
````

Primero eliminamos aquellas filas que poseen información de las marcas que no utilizaremos, concretamente `Others` y `Total_Market`

````python
shipments_in_mn.drop([5,6], inplace = True)
````

Ahora graficaremos nuestro datos en un grafico de barras, el cual mostrara las ventas de cada marca por los trimestre de cada año. Para esto es necesario "pivotear" la tabla, utilizaremos el método de Pandas `DataFrame.melt()`

````python
shipments_in_mn = shipments_in_mn.melt(id_vars=["Brands"], var_name="Quarter", value_name = "Sold")
````

````python
shipments_in_mn.head()
````

````python
shipments_in_mn.info()
````

````python
shipments_in_mn['Quarter'] = shipments_in_mn['Quarter'].str.replace(r'(Q\d)_(\d+)', r'\2\1', regex= True)
shipments_in_mn['date'] = pd.PeriodIndex(shipments_in_mn['Quarter'], freq = "Q").to_timestamp()
shipments_in_mn['year'] = pd.to_datetime(shipments_in_mn["date"], format = "%Y-%m-%d")
shipments_in_mn['year'] = shipments_in_mn['year'].dt.year
shipments_in_mn.head()
````

````python
(ggplot(shipments_in_mn)+
 aes(x="Quarter",y = "Sold" , fill="Brands")+
 geom_bar(stat= "identity", position="dodge2")+
 coord_flip()+
  theme_dark()+
 facet_wrap("Brands")+
 ggtitle("Ventas trimestrales en millones de dolares")+
 xlab("Trimestres")+
  ylab("Millones de dolares")+
  scale_fill_manual(name="Marca",values=["lightgray","springgreen","deepskyblue","powderblue","darkorange"])
)
````

````python
(ggplot(shipments_in_mn)+
 aes(x = "Brands", y = "Sold", fill = "Brands") +
 geom_bar(stat = "identity") +
 facet_wrap("year") +
 theme_dark()+
 theme(axis_text_x=element_text(rotation=45, hjust=1))+
 ggtitle("Ventas por año de las marcas en millones de dolares")+
 ylab("Millones de dolares")+
 xlab("marcas")+
 scale_x_discrete(labels = [""], breaks = [""]) + labs(x = "")+
 scale_fill_manual(name="Marca",values=["lightgray","springgreen","deepskyblue","powderblue","darkorange"])
  )
````

Al igual que con el dataset anterior, descartaremos aquella información que no utilizaremos, para esto conservaremos las siguientes columnas:
- Marca
- Popularidad
- Precio
- Fecha de lanzamiento

````python
# Descartar columnas que no se utilizarán
phone_data = phone_data[["brand_name","popularity","best_price","release_date"]]
phone_data
````

````python
phone_data['release_date'] = pd.to_datetime(phone_data["release_date"], format = "%m-%Y")
phone_data['year'] = phone_data['release_date'].dt.year
phone_data.head()
````

Este dataset a diferencia del anterior posee información de otras marcas tales como
nokia, alcatel entre otras. Como no poseemos mayor información de estas nos quedaremos solo
con aquellos celulares que sean de las marcas anteriormente mencionadas.

````python
to_keep = ['Oppo', 'Samsung', 'Apple', 'Xiaomi', 'Vivo']
phone_data = phone_data[phone_data['brand_name'].isin(to_keep)]
phone_data
````

Transformamos de Grivna(divisa Ucraniana) a dólares (Valor actualizado al 2022-05-15)

````python
phone_data["best_price"] = phone_data[["best_price"]].apply(lambda row: row * 0.034 , axis=1)
````

Posteriormente asignaremos cada uno de estos dispositivos a un trimestre correspondiente


````python
phone_data["release_date"] = pd.to_datetime(phone_data['release_date'])
phone_data['quarter'] = pd.PeriodIndex(phone_data["release_date"], freq = "Q")
phone_data
````

````python
phone_data = phone_data[(phone_data["year"]==2019) | (phone_data["year"]==2020) | (phone_data["year"]==2021) ]
````

Ahora categorizaremos los dispositivos respecto a su valor de venta. Dividiendo en un precio bajo, medio y alto.

````python
bins = list(range(0, 1900+1, int(1900/3)))
phone_data['price_group'] = pd.cut(phone_data["best_price"], bins = bins)
phone_data
````

#Analisis por  la popularidad de las marcas

- Analizaremos la popularidad por rango de precio de sus "smartpone"
- Histograma de Precios


````python
(ggplot(phone_data)+
 aes(x = "brand_name", y = "popularity", fill = "brand_name") +
 geom_bar(stat = "summary") +
 facet_wrap("year") +
 theme_dark()+
 theme(axis_text_x=element_text(rotation=45, hjust=1))+
 ggtitle("Popularidad de las marcas por años")+
 ylab("Popularidad")+
 xlab("Marcas")+
 scale_x_discrete(labels = [""], breaks = [""]) + labs(x = "")+
 scale_fill_manual(name="Marca",values=["lightgray","springgreen","deepskyblue","powderblue","darkorange"])
  )
````

````python
(
    ggplot(phone_data, aes(x = "brand_name", y="popularity", fill = "brand_name")) +
    geom_bar(position = "dodge", stat = "summary") + facet_wrap("price_group") +
    theme_dark()+
    theme(axis_text_x=element_text(rotation=45, hjust=1))+
    ggtitle("Rango de precios de equipos por Popularidad")+
    xlab("Marcas")+
    ylab("Popularidad")+
    scale_x_discrete(labels = [""], breaks = [""]) + labs(x = "")+
    scale_fill_manual(name="Marca",values=["lightgray","springgreen","deepskyblue","powderblue","darkorange"])
)
````
