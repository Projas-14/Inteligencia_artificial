Paso 3: Proyección de Estados Financieros
Utilizaremos técnicas de modelado financiero para proyectar los estados financieros futuros de la empresa, como el estado de resultados proyectado, el balance general proyectado y el estado de flujo de efectivo proyectado.

Implementaremos un modelo financiero simple para proyectar los ingresos y los gastos futuros de la empresa y calcular las utilidades netas proyectadas. Utilizaremos un enfoque de proyección de ingresos basado en el crecimiento histórico y un enfoque de proyección de gastos basado en porcentajes de los ingresos proyectados.

```python
# Implementación del modelo financiero para proyecciones

# Número de acciones en circulación (puedes obtener esta información de fuentes financieras)
numero_de_acciones = 1000000  # Por ejemplo, 1 millón de acciones

# Proyección de ingresos futuros
tasa_crecimiento = 0.05  # Tasa de crecimiento esperada (por ejemplo, 5%)
ingresos_actuales = datos_historicos['Close'][-1] * numero_de_acciones  # Ingresos actuales (precio de cierre actual * número de acciones en circulación)
anos_a_proyectar = 5  # Número de años a proyectar

proyeccion_ingresos = [ingresos_actuales * (1 + tasa_crecimiento) ** i for i in range(1, anos_a_proyectar+1)]

# Proyección de gastos futuros (supondremos un porcentaje constante de los ingresos proyectados)
porcentaje_gastos = 0.6  # Porcentaje de gastos respecto a los ingresos
gastos_proyectados = [ingresos * porcentaje_gastos for ingresos in proyeccion_ingresos]

# Calculo de utilidades netas proyectadas
utilidades_netas_proyectadas = [ingresos - gastos for ingresos, gastos in zip(proyeccion_ingresos, gastos_proyectados)]
```

Paso 5: Sensibilidad y Análisis de Escenarios
Realizamos análisis de sensibilidad y escenarios para evaluar cómo diferentes variables y suposiciones pueden afectar las proyecciones de utilidades futuras.
Este paso puede incluir la realización de análisis de sensibilidad sobre la tasa de crecimiento esperada y el porcentaje de gastos, así como la evaluación de diferentes escenarios. A continuación, se muestra un ejemplo básico de cómo puedes realizar un análisis de sensibilidad sobre la tasa de crecimiento:

```python
# Análisis de sensibilidad sobre la tasa de crecimiento
tasas_crecimiento_sensibilidad = [0.04, 0.05, 0.06]  # Valores de tasa de crecimiento para probar

resultados_sensibilidad = {}

for tasa in tasas_crecimiento_sensibilidad:
    proyeccion_ingresos = [ingresos_actuales * (1 + tasa) ** i for i in range(1, anos_a_proyectar+1)]
    gastos_proyectados = [ingresos * porcentaje_gastos for ingresos in proyeccion_ingresos]
    utilidades_netas_proyectadas = [ingresos - gastos for ingresos, gastos in zip(proyeccion_ingresos, gastos_proyectados)]
    resultados_sensibilidad[tasa] = utilidades_netas_proyectadas
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los resultados
df_resultados = pd.DataFrame({
    'Año': range(1, anos_a_proyectar + 1),
    'Ingresos Proyectados': proyeccion_ingresos,
    'Gastos Proyectados': gastos_proyectados,
    'Utilidades Netas Proyectadas': utilidades_netas_proyectadas
})

# Gráfico de línea para mostrar los ingresos proyectados y los gastos proyectados
plt.figure(figsize=(10, 6))
plt.plot(df_resultados['Año'], df_resultados['Ingresos Proyectados'], label='Ingresos Proyectados', marker='o')
plt.plot(df_resultados['Año'], df_resultados['Gastos Proyectados'], label='Gastos Proyectados', marker='o')
plt.title('Proyección de Ingresos y Gastos')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()

# Tabla para mostrar las utilidades netas proyectadas
print("Utilidades Netas Proyectadas:")
print(df_resultados[['Año', 'Utilidades Netas Proyectadas']])
```

Para calcular de manera dinámica la tasa de crecimiento y las utilidades netas cada vez que ingresas un nuevo dato del mes, podrías seguir los siguientes pasos:

Mantén un registro de los datos históricos de los ingresos de cada mes.
Calcula la tasa de crecimiento entre los ingresos del mes actual y el mes anterior para obtener una estimación de la tasa de crecimiento.
Utiliza la tasa de crecimiento estimada para proyectar los ingresos y los gastos para el próximo mes.
Calcula las utilidades netas proyectadas para el próximo mes.
Actualiza tus proyecciones y registros con los nuevos datos del mes.

```python
import matplotlib.pyplot as plt

# Supongamos que tienes una lista de ingresos mensuales
ingresos_mensuales = [10000, 12000, 14000, 16000, 18000]  # Ejemplo de ingresos mensuales para los últimos 5 meses

# Calcula la tasa de crecimiento entre el mes actual y el mes anterior
tasa_crecimiento_actual = (ingresos_mensuales[-1] - ingresos_mensuales[-2]) / ingresos_mensuales[-2]

# Proyecta los ingresos y los gastos para el próximo mes
ingresos_proyectados_siguiente_mes = ingresos_mensuales[-1] * (1 + tasa_crecimiento_actual)
porcentaje_gastos = 0.6  # Porcentaje de gastos respecto a los ingresos
gastos_proyectados_siguiente_mes = ingresos_proyectados_siguiente_mes * porcentaje_gastos

# Calcula las utilidades netas proyectadas para el próximo mes
utilidades_netas_proyectadas_siguiente_mes = ingresos_proyectados_siguiente_mes - gastos_proyectados_siguiente_mes

# Definir el nuevo ingreso del mes
nuevo_ingreso_del_mes = 20000  # Por ejemplo, el ingreso del mes actual es de $20,000

# Actualizar los registros con el nuevo dato del mes
ingresos_mensuales.append(nuevo_ingreso_del_mes)

# Agregar el nombre del próximo mes a la lista de meses
meses = [f"Mes {i}" for i in range(1, len(ingresos_mensuales) + 1)]
meses.append("Próximo Mes")

# Agregar el ingreso proyectado del próximo mes a la lista de ingresos
ingresos = ingresos_mensuales + [ingresos_proyectados_siguiente_mes]

# Presentar los resultados de manera comprensible
# Mostrar los ingresos mensuales históricos en una tabla
print("Ingresos Mensuales Históricos:")
for i, ingreso in enumerate(ingresos_mensuales, start=1):
    print(f"Mes {i}: {ingreso}")

# Mostrar la tasa de crecimiento estimada y las utilidades netas proyectadas para el próximo mes
print("\nResultados Proyectados para el Próximo Mes:")
print(f"Tasa de Crecimiento Estimada: {tasa_crecimiento_actual:.2%}")
print(f"Ingresos Proyectados para el Próximo Mes: {ingresos_proyectados_siguiente_mes}")
print(f"Gastos Proyectados para el Próximo Mes: {gastos_proyectados_siguiente_mes}")
print(f"Utilidades Netas Proyectadas para el Próximo Mes: {utilidades_netas_proyectadas_siguiente_mes}")

# Mostrar un gráfico de barras para visualizar los ingresos mensuales históricos y el ingreso proyectado para el próximo mes
plt.figure(figsize=(10, 6))
plt.bar(meses, ingresos, color='skyblue')
plt.title('Ingresos Mensuales Históricos y Proyectados')
plt.xlabel('Mes')
plt.ylabel('Ingresos')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Seleccionar solo la columna 'Close' (precio de cierre)
closing_prices = datos_historicos['Close']

# Reiniciar el índice para tener un índice numérico
closing_prices.reset_index(drop=True, inplace=True)

# Crear una variable de tiempo numérica para usar como característica en el modelo
time_index = pd.Series(range(len(closing_prices)))

# Añadir la variable de tiempo al DataFrame
datos_historicos['Time'] = time_index

# Mostrar las primeras filas del DataFrame modificado
print(datos_historicos.head())
```

Paso 3: Ajustar un modelo de regresión lineal y predecir las tendencias futuras

```python
# Importar las bibliotecas necesarias
import pandas as pd
import pandas_datareader as pdr

# Obtener datos históricos hasta la fecha actual
empresa = 'AAPL'
datos_historicos = yf.download(empresa, start='2010-01-01', end='2023-12-31')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Eliminar la columna 'Adj Close' (ajuste de cierre) si existe
if 'Adj Close' in datos_historicos.columns:
    datos_historicos.drop(columns=['Adj Close'], inplace=True)

# Imputar valores faltantes usando la mediana de los datos existentes
imputer = SimpleImputer(strategy='median')
datos_historicos_imputed = pd.DataFrame(imputer.fit_transform(datos_historicos), columns=datos_historicos.columns)

# Verificar si la columna 'Time' está presente en los datos
if 'Time' not in datos_historicos_imputed.columns:
    # Si no está presente, generamos una nueva columna de tiempo
    datos_historicos_imputed['Time'] = range(len(datos_historicos_imputed))

# Seleccionar solo la columna 'Close' (precio de cierre) y la columna 'Time' (tiempo)
features = datos_historicos_imputed[['Time']]
target = datos_historicos_imputed['Close']

# Crear el objeto de modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo utilizando las variables de tiempo y precios de cierre
model.fit(features, target)

# Predecir los precios de cierre para los próximos años (2024-2027)
future_years = range(2024, 2028)
future_time_index = pd.Series(range(len(datos_historicos_imputed), len(datos_historicos_imputed) + len(future_years)))
future_prices = model.predict(future_time_index.values.reshape(-1, 1))

# Mostrar las predicciones de precios de cierre para los próximos años
print("Predicciones de precios de cierre para los próximos años:")
for year, price in zip(future_years, future_prices):
    print(f"Año {year}: ${price:.2f}")
```
